```python
import pandas as pd
```

```python
node = pd.read_csv("nodes.csv")
```

```python
node.head(5)
```

```python
node.info()
```

```python
nodes_dict = {
    item["link"]: {
        "name" : item["name"],
        "type" : item["type"]
    } for _, item in node.iterrows()
}
```

```python
edges = pd.read_csv("edges.csv")
```

```python
edges.info()
```

```python
s = set()

for _, edge in edges.iterrows():
    # print(edge["src"], edge["des"], edge["type"])
    try:
        s.add((nodes_dict[edge["src"]]["type"], edge["type"], nodes_dict[edge["des"]]["type"]))
    except Exception as e:
        print(e)
```

```python
print(f"Len: {len(s)}")

with open("relationship.txt", "w", encoding="utf-8") as f:
    for item in s:
        f.write(f"({item[0]}, {item[1]}, {item[2]})\n")
```
